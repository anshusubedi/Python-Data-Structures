Project: Retail Sales Analysis

We will use Matplotlib to visualize data for a project. The objective is to analyze sales performance for the year. The dataset includes monthly sales data, sales by product category, and customer satisfaction ratings. We'll create various plots to visualize this data.

```python
#Line Plot
import matplotlib.pyplot as plt

# Data: Monthly sales in thousands of dollars
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [50, 55, 60, 58, 65, 70, 75, 72, 80, 85, 90, 100]

# Create the plot
plt.plot(months, sales, marker='o', color='purple', linestyle='-', linewidth=2, markersize= 10)

# Add titles and labels
plt.title('Monthly Sales Performance (2023)')
plt.xlabel('Month')
plt.ylabel('Sales (in $1000s)')
plt.grid(True)

#  Highlight the highest sales month
max_sales = max(sales)
max_month = months[sales.index(max_sales)]
plt.annotate(f'Highest Sales: {max_sales}', xy=(max_month, max_sales), xytext=(max_month, max_sales+10),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Show the plot
plt.show()
```

```python
#Bar Chart: Sales by Product Category

#Next, we'll visualize sales by different product categories to identify which categories are driving the most revenue.

# Data: Sales by product category (in thousands of dollars)
categories = ['Electronics', 'Clothing', 'Home Goods', 'Beauty', 'Toys']
category_sales = [150, 130, 120, 80, 90]

# Create the bar chart
plt.bar(categories, category_sales, color=['blue', 'green', 'red', 'purple', 'orange'])

# Add titles and labels
plt.title('Sales by Product Category (2023)')
plt.xlabel('Product Category')
plt.ylabel('Sales (in $1000s)')

# Highlight the best-selling category
max_sales_category = max(category_sales)
best_category = categories[category_sales.index(max_sales_category)]
plt.annotate(f'Best-Selling Category: {best_category}', xy=(best_category, max_sales_category),
             xytext=(best_category, max_sales_category + 10),
             arrowprops=dict(facecolor='black', shrink=0.03))
```

```python
#Pie Chart: Customer Satisfaction Distribution

#We'll now look at customer satisfaction ratings to understand how customers feel about their purchases.

import matplotlib.pyplot as plt

# Data: Customer satisfaction ratings distribution
satisfaction_labels = ['Very Satisfied', 'Satisfied', 'Neutral', 'Dissatisfied', 'Very Dissatisfied']
satisfaction_counts = [400, 300, 150, 100, 50]

# Create the pie chart
plt.pie(satisfaction_counts, labels=satisfaction_labels, autopct='%1.1f%%', colors=['green', 'lightgreen', 'pink', 'orange', 'grey'], startangle=140)

# Add title
plt.title('Customer Satisfaction Ratings (2023)')

# Show the plot
plt.show()
```

```python
#Stacked Bar Chart: Sales and Profit by Quarter

#Finally, let's compare sales and profit across different quarters of the year.

import matplotlib.pyplot as plt

# Data: Sales and Profit by quarter (in thousands of dollars)
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
sales = [165, 193, 227, 275]
profit = [45, 50, 60, 75]

# Create stacked bar chart
plt.bar(quarters, sales, label='Sales', color='blue')
plt.bar(quarters, profit, label='Profit', color='green', bottom=sales)

# Add titles and labels
plt.title('Sales and Profit by Quarter (2023)')
plt.xlabel('Quarter')
plt.ylabel('Amount (in $1000s)')

# Add legend
plt.legend()

# Show the plot
plt.show()

```

```python
#Heatmap: Monthly Sales Performance by Region

#We'll create a heatmap to analyze how different regions performed each month.

import matplotlib.pyplot as plt
import numpy as np

# Data: Sales performance by region (in thousands of dollars)
regions = ['North', 'South', 'East', 'West']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_data = np.array([
    [15, 18, 20, 25, 22, 28, 30, 32, 35, 38, 42, 45],
    [20, 22, 25, 28, 30, 35, 38, 40, 42, 45, 48, 50],
    [10, 12, 15, 18, 20, 22, 25, 27, 30, 32, 35, 38],
    [18, 20, 22, 25, 28, 30, 32, 35, 38, 40, 42, 45]
])

# Create heatmap
plt.imshow(sales_data, cmap='Blues', interpolation='nearest')

# Add titles and labels
plt.title('Monthly Sales Performance by Region (2023)')
plt.xlabel('Month')
plt.ylabel('Region')

# Set ticks
plt.xticks(ticks=np.arange(len(months)), labels=months)
plt.yticks(ticks=np.arange(len(regions)), labels=regions)

# Annotate each cell with the value
for i in range(len(regions)):
    for j in range(len(months)):
        plt.text(j, i, f'{sales_data[i, j]}', ha='center', va='center', color='black')

# Show color bar
plt.colorbar()

# Show the plot
plt.show()

```
